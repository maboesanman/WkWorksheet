# WaniKani Cache Exploration

This notebook demonstrates how to use the WaniKani caching layer.

```python
# Import the cache module
from wkworksheet.wanikani_cache import WaniKaniCache

# Initialize cache and fetch subjects (idempotent)
cache = WaniKaniCache()
cache.fetch_subjects()

print("Cache loaded!")
```

```python
# Get all kanji subjects
kanji = cache.get_subjects("kanji")
radicals = cache.get_subjects("radical")
vocabulary = cache.get_subjects("vocabulary")

print(f"Total subjects: {len(cache.get_subjects())}")
print(f"  Kanji: {len(kanji)}")
print(f"  Radicals: {len(radicals)}")
print(f"  Vocabulary: {len(vocabulary)}")
```

```python
# Explore a specific kanji
first_kanji = kanji[0]
print(f"Example kanji: {first_kanji['data']['characters']}")
print(f"  Level: {first_kanji['data']['level']}")
print(f"  Meanings: {[m['meaning'] for m in first_kanji['data']['meanings'] if m['primary']]}")
print(f"  Readings: {first_kanji['data']['readings']}")
```

```python
# Find all level 1 kanji
level_1_kanji = [k for k in kanji if k['data']['level'] == 1]

print(f"\nLevel 1 kanji ({len(level_1_kanji)} total):")
for k in level_1_kanji:
    data = k['data']
    meanings = ', '.join(m['meaning'] for m in data['meanings'] if m['primary'])
    print(f"  {data['characters']} - {meanings}")
```

```python
# Count kanji by level
from collections import Counter

level_counts = Counter(k['data']['level'] for k in kanji)

print("Kanji distribution by level:")
for level in sorted(level_counts.keys())[:10]:  # Show first 10 levels
    print(f"  Level {level:2d}: {level_counts[level]:3d} kanji")
```

```python
# Play with filtering by readings
# Find kanji with reading containing "か"
kanji_with_ka = [
    k for k in kanji
    if any('か' in r['reading'] for r in k['data']['readings'])
]

print(f"\nKanji with 'か' in reading: {len(kanji_with_ka)}")
for k in kanji_with_ka[:5]:  # Show first 5
    data = k['data']
    readings = ', '.join(r['reading'] for r in data['readings'] if r['primary'])
    meanings = ', '.join(m['meaning'] for m in data['meanings'] if m['primary'])
    print(f"  {data['characters']} - {readings} - {meanings}")
```
