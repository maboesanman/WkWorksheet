# WaniKani Cache Exploration

This notebook demonstrates how to use the WaniKani caching layer.

```python
# Import the cache module
from wkworksheet.wanikani_cache import WaniKaniCache

# Initialize cache and fetch subjects (idempotent)
cache = WaniKaniCache()
# cache.fetch_subjects()
# cache.fetch_review_statistics()
# cache.fetch_assignments()
# cache.fetch_user()

print("Cache loaded!")
```

```python
# Get all subjects
kanji = cache.get_subjects("kanji")
radicals = cache.get_subjects("radical")
vocabulary = cache.get_subjects("vocabulary")

print(f"Total subjects: {len(cache.get_subjects())}")
print(f"  Kanji: {len(kanji)}")
print(f"  Radicals: {len(radicals)}")
print(f"  Vocabulary: {len(vocabulary)}")
```

```python
# Find all level 1 kanji
level_1_kanji = [k for k in kanji if k['data']['level'] == 10]

print(f"\nLevel 1 kanji ({len(level_1_kanji)} total):")
for k in level_1_kanji:
    data = k['data']
    meanings = ', '.join(m['meaning'] for m in data['meanings'] if m['primary'])
    print(f"  {data['characters']} - {meanings} - {k['id']}")
```
